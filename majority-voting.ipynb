```python
import pandas as pd
import os
SUBS_DIR = 'subs'
```

```python
df = pd.read_csv(os.path.join(SUBS_DIR, 'seresnet-8a.csv'), index_col='fname')
df.sort_index(inplace=True)
subs_list = [
    'seresnet-8a-tta-2.csv',
    'seresnet-8a-tta-1.csv',
    'seresnet-4-best-no-tta.csv',
    'seresnet-8a-best.csv',
    'seresnet-4-tta-2.csv',
    'seresnet-4-best-tta-4.csv',
    'seresnet-4-best-tta-4-2.csv',
    'seresnet-8.csv',
    'seresnet-8-best.csv',
    'seresnet-7a.csv',
    'seresnet-1a.csv',
    'seresnet-7a-best.csv',
    'seresnet-6-best.csv',
    'seresnet-1a-best.csv',
    'seresnet-2-best.csv',
    'seresnet-1-best.csv',
    'seresnet-7-best.csv',
    'seresnet-3-best.csv',
]
```

```python
for sub in subs_list:
    sub_df = pd.read_csv(os.path.join(SUBS_DIR, sub), index_col='fname')
    sub_df.sort_index(inplace=True)
    df[sub] = sub_df['label']
```

```python
df.head()
```

```python
from tqdm import tqdm
labels = []
for i in tqdm(range(len(df))):
    labels.append(df.iloc[i].value_counts().idxmax())
```

```python
data = {
    'fname': df.index,
    'label': labels
}
ensemble = pd.DataFrame(data)
```

```python
ensemble.to_csv(os.path.join(SUBS_DIR, 'ensemble-all-86+.csv'), index=False)
```
